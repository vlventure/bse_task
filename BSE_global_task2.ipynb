```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
pre_post_df = pd.read_csv('bse_pre_post_data.csv')
```

```python
pre_post_df.head()
```

```python
pre_post_df['Tenure'].value_counts()
```

I noticed that there was a category from 11-20 which overlapped with other categories.  
I decided to get rid of any rows that had a tenure of '11-20 season' because I did not know if those belonged to the 11-15 category or 16-20 category.

```python
#keep only rows where tenure is not equal to '11-20 season'
pre_post_df = pre_post_df[pre_post_df.Tenure != '11-20 season']
```

```python
pre_post_df['Tenure'].value_counts()
```

I noticed that there were some phrases that had lower case letters and some that had uppercase letters. I wanted to replace the phrases with lower case letters so that they all matched.

```python
#the following code is used to replace strings where the second word starts with a lower case letter
pre_post_df.loc['Comm_Freq'] = pre_post_df.loc['Comm_Freq'].str.replace('Too few', 'Too Few')
pre_post_df.loc['Comm_Freq'] = pre_post_df.loc['Comm_Freq'].str.replace('Too many', 'Too Many')
pre_post_df.loc['Comm_Freq'] = pre_post_df.loc['Comm_Freq'].str.replace('Just right', 'Just Right')
```
